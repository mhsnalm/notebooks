# Execute message flow

```python
url = base_url + "/ExecuteIntegration/"

payload = json.dumps({
  "MessageFlowName": "GenericSetupExport",
  "SubTenantId": 1,
  "Params": {
    "screenid": 190
  }
})
headers = {
  '_context': auth_token,
  'Content-Type': 'application/json'
}

response = requests.request("POST", url, headers=headers, data=payload)
response_json = json.loads(response.text);


f = open("excel.xlsx", "wb")
f.write(bytearray(base64.b64decode(response_json['Data']['Params']['xlsx-mapforce.xlsx'])))
f.close()

excel_data_fragment = pandas.read_excel('excel.xlsx')

json_data = excel_data_fragment.to_json() #excel2json.convert_from_file('excel.xls')

print(json_data)
#print(json.dumps(response_json['Data']['Params']['xlsx-mapforce.xlsx'], sort_keys=True, indent=4))
```
